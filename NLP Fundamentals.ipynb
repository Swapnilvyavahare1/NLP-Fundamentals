```python
import nltk
nltk.download()
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize
```

```python
my_paragraph= 'Defence Minister Rajnath Singh’s message to his visiting Chinese counterpart General Li Shangfu, that Beijing’s violation of border agreements had “eroded the entire basis of bilateral relations”, has once again served as a reminder of how the two neighbours remain far apart in their assessments of what ails their relationship, and how to fix it. Thursday’s talks, on the sidelines of the Shanghai Cooperation Organisation Defence Ministers Meeting in New Delhi, saw both sides reiterate their respective stands on the border issue but no meeting of minds. The Defence Minister underlined India’s position that the development of ties is premised on peace on the borders. While India has continued to convey a sense of urgency to resolve the Line of Actual Control (LAC) crisis, the Chinese Defence Minister, on the other hand, called on India to “take a long-term view” and “place the border issue in an appropriate position in bilateral relations”, a divergence from India’s stand that the rest of relations is predicated on peace along the LAC. Rajnath Singh conveyed to Beijing that if normalcy is to be restored in ties, disengagement in the two remaining friction points will need to be followed by de-escalation. This includes the eventual de-induction of the estimated one lakh troops from both sides that have remained deployed in forward areas for close to three years — a situation not seen along the India-China border in more than three decades.'
```

```python
my_paragraph
```

# Sentence tokenization

```python
sent_tokenize(my_paragraph)
```

```python
for sent in sent_tokenize(my_paragraph):
    print(sent)
```

# Word tokenization

```python
word_tokenize(my_paragraph)
```

```python
for word in word_tokenize(my_paragraph):
    print(word)
```

# Whitespace tokenization

It removes symbols and converts sentences into words.

```python
white_tokenize=nltk.tokenize.WhitespaceTokenizer()
```

```python
white_tokenize.tokenize(my_paragraph)
```

```python
for whitespace in white_tokenize.tokenize(my_paragraph):
    print(whitespace)
```

```python
word_punctuation.tokenize(complete_sentence)
```

```python
# Stopwords

stop_words=set(stopwords.words('english'))
```

```python
sent_filter=[w for w in complete_sentence if not w in stop_words]
```

```python
sent_filter
```

```python
# To visualize while applying stopwords

from wordcloud import WordCloud,STOPWORDS
```

```python
word_stop=nltk.corpus.stopwords.words('english')
```

```python
word_stop
```

```python
wcloud=WordCloud(max_words=50,stopwords=word_stop)
wcloud
```

```python
wcloud.generate(complete_sentence)

import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
plt.imshow(wcloud)
plt.show()
```

```python
wcloud=WordCloud(max_words=50,stopwords=word_stop+['road'])
```

```python
wcloud.generate(complete_sentence)

import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
plt.imshow(wcloud)
plt.show()
```

```python
wcloud=WordCloud(max_words=50,stopwords=word_stop+['road'],background_color='white')

wcloud.generate(complete_sentence)

import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
plt.imshow(wcloud)
plt.show()
```
